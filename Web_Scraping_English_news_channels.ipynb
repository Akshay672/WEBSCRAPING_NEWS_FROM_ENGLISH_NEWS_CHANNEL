# ENGLISH NEWS CHANNEL

```python
import requests
from bs4 import BeautifulSoup
```

### Single category

## Times of india (English)

```python
url = "https://timesofindia.indiatimes.com/briefs"
page_request = requests.get(url)
data = page_request.content
soup = BeautifulSoup(data,"html.parser")
```

```python
for divtag in soup.find_all('div', {'class': 'brief_box'}):
    #print(divtag)
    try:
        print('News link:')
        print("https://timesofindia.indiatimes.com" + divtag.find('a')['href'])#link
        print('')
        print('News Category:')
        print(divtag.find('span').text)#category
        print('')
        print('News Title:')
        print(divtag.find('h2').text)#title
        print('')
        print('Image Link:')
        print(divtag.find('img')['data-src'])#image
        print('')
        print('News Description:')
        print(divtag.find('p').text)#description
        print('------------------------------------------------------------------------------------------------------')
    except:
        continue
```
